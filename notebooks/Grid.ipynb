```python
from sklearn.gaussian_process import GaussianProcessRegressor
from sklearn.gaussian_process.kernels import Matern
```

```python
X = grid[line_cols].values
y_col = 'log U'
regressor = GaussianProcessRegressor(n_restarts_optimizer=10, normalize_y=True, kernel=Matern(), alpha=0.5)  # RBF smooth k
# regressor = GaussianProcessRegressor(n_restarts_optimizer=10, normalize_y=True)  # RBF smooth k
y = grid[y_col].values
regressor.fit(X, y)  # set up to predict that column
```

```python
# y_est, std = regressor.predict(X, return_std=True)
y_est = regressor.predict(X)
# # y_est = regressor.sample_y(X)
# y_est = y_est.squeeze()
```

```python
regressor.score(X, y)
```
